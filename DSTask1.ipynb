## **TASK 1**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("worldpopulationdata.csv")
df
```

```python
df.columns
```

```python
df.drop(['Country Name','Indicator Name'],axis=1,inplace=True)
df
```

##**DATA OF TOP 10 COUNTRIES WITH HIGHEST POPULATION**

```python
totalpopdata = df[df['Indicator Code']=='SP.POP.TOTL']
totaldata = totalpopdata.sort_values(by="2022", ascending = False)
totaldata22 = totaldata.head(10)
totaldata22
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=totaldata22, palette="rocket")
plt.xlabel("Total population")
plt.ylabel("Country")
plt.title("Top 10 countries with highest population in 2022")


plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2012",y="Country Code",data=totaldata22, palette="cubehelix")
plt.xlabel("Total population")
plt.ylabel("Country")
plt.title("Top 10 countries with highest population in 2012")
```

## **DATA OF TOP 10 COUNTRIES WITH LOWEST POPULATION**

```python
totalpopdata = df[df['Indicator Code']=='SP.POP.TOTL']
totaldata = totalpopdata.sort_values(by="2022", ascending = True)
totaldata22 = totaldata.head(10)
totaldata22
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=totaldata22, palette="rocket")
plt.xlabel("Total population")
plt.ylabel("Country")
plt.title("Top 10 countries with lowest population in 2022")


plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2012",y="Country Code",data=totaldata22, palette="cubehelix")
plt.xlabel("Total population")
plt.ylabel("Country")
plt.title("Top 10 countries with lowest population in 2012")
```

## **TOP 10 COUNTRIES WITH HIGHEST MALE POPULATION**

```python
malepopdata = df[df['Indicator Code']=='SP.POP.TOTL.MA.IN']
maledata = malepopdata.sort_values(by="2022", ascending = False)
maledata22 = maledata.head(10)
maledata22
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=maledata22, palette="mako")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.title("Top 10 countries with highest male population in 2022")
```

## **TOP 10 COUNTRIES WITH HIGHEST FEMALE POPULATION**

```python
femalepopdata = df[df['Indicator Code']== 'SP.POP.TOTL.FE.IN']
femaledata = femalepopdata.sort_values(by="2022", ascending = False)
femaledata22 = femaledata.head(10)
femaledata22
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data=femaledata22, palette="rocket")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.title("Top 10 countries with highest female population in 2022")
```

## **COMPARITIVE PLOT FOR BOTH MALE AND FEMALE IN 2022**

```python
mergesets = pd.merge(maledata22,femaledata22,on="Country Code")
mergesets["2022"] = mergesets["2022_x"] - mergesets["2022_y"]
mergesets
```

```python
sorted_data = mergesets.sort_values(by="2022", ascending=False)
topcnt = sorted_data.head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y= "2022_x",data=topcnt, color="purple",label="female")
sns.barplot(x="Country Code",y= "2022_y",data=topcnt, bottom = topcnt["2022_y"],color="black", label = "male")
plt.xlabel("Country")
plt.ylabel("Total Population")
plt.title("Male and Female population in 2022")
```
